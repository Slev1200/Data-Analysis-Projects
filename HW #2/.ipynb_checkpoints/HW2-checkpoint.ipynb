```python
# This must be the only cell you read the data. Do not read data in other cells.

from matplotlib import pyplot as plt
import csv
import json
import numpy as np
from scipy.stats import norm
import random

def jprint(data):
    print(json.dumps(data, indent=4))

data=[]
counter=0
with open('input.csv', encoding="utf-8") as csv_file:
    csv_reader = csv.DictReader(csv_file)
    for row in csv_reader:
        row["Temperature(F)"]=float(row["Temperature(F)"])
        row["Humidity(%)"]=float(row["Humidity(%)"])
        row["Pressure(in)"]=float(row["Pressure(in)"])
        row["Visibility(mi)"]=float(row["Visibility(mi)"])
        row["Severity"]=int(row["Severity"])
        data.append(row)
#jprint(data)
test=[]
with open('test.csv', encoding="utf-8") as csv_file:
    csv_reader = csv.DictReader(csv_file)
    for row in csv_reader:
        test.append(row)
```

```python
# Q1
attr = ["US/Central", "US/Eastern", "US/Mountain", "US/Pacific"]
temp1 = [i["Temperature(F)"] for i in data if i["Timezone"] == "US/Pacific"]
temp2 = [i["Temperature(F)"] for i in data if i["Timezone"] == "US/Eastern"]
temp3 = [i["Temperature(F)"] for i in data if i["Timezone"] == "US/Mountain"]
temp4 = [i["Temperature(F)"] for i in data if i["Timezone"] == "US/Central"]

humidity1 = [i["Humidity(%)"] for i in data if i["Timezone"] == "US/Pacific"]
humidity2 = [i["Humidity(%)"] for i in data if i["Timezone"] == "US/Eastern"]
humidity3 = [i["Humidity(%)"] for i in data if i["Timezone"] == "US/Mountain"]
humidity4 = [i["Humidity(%)"] for i in data if i["Timezone"] == "US/Central"]

pressure1 = [i["Pressure(in)"] for i in data if i["Timezone"] == "US/Pacific"]
pressure2 = [i["Pressure(in)"] for i in data if i["Timezone"] == "US/Eastern"]
pressure3 = [i["Pressure(in)"] for i in data if i["Timezone"] == "US/Mountain"]
pressure4 = [i["Pressure(in)"] for i in data if i["Timezone"] == "US/Central"]

visibility1 = [i["Visibility(mi)"] for i in data if i["Timezone"] == "US/Pacific"]
visibility2 = [i["Visibility(mi)"] for i in data if i["Timezone"] == "US/Eastern"]
visibility3 = [i["Visibility(mi)"] for i in data if i["Timezone"] == "US/Mountain"]
visibility4 = [i["Visibility(mi)"] for i in data if i["Timezone"] == "US/Central"]

plt.figure(figsize = (20, 4))
plt.subplot(1, 4, 1)
means = [np.mean(temp1), np.mean(temp2), np.mean(temp3), np.mean(temp4)]
stds = [np.std(temp1, ddof = 1), 
        np.std(temp2, ddof = 1), 
        np.std(temp3, ddof = 1), 
        np.std(temp4, ddof = 1)]
plt.bar(attr, means, yerr = stds, color = 'darkorange', capsize = 10, width = 0.3)
plt.xticks()

plt.ylabel ("Temperature(F)")

plt.subplot(1, 4, 2)
means2 = [np.mean(humidity1), np.mean(humidity1), np.mean(humidity1), np.mean(humidity1)]
stds2 = [np.std(humidity1, ddof = 1), 
        np.std(humidity1, ddof = 1), 
        np.std(humidity1, ddof = 1), 
        np.std(humidity1, ddof = 1)]
plt.bar(attr, means2, yerr = stds2, color = 'green', capsize = 10, width = 0.3)
plt.xticks()

plt.ylabel ("Humidity(%)")

plt.subplot(1, 4, 3)
means3 = [np.mean(pressure1), np.mean(pressure2), np.mean(pressure3), np.mean(pressure4)]
stds3 = [np.std(pressure1, ddof = 1), 
        np.std(pressure2, ddof = 1), 
        np.std(pressure3, ddof = 1), 
        np.std(pressure4, ddof = 1)]
plt.bar(attr, means3, yerr = stds3, color = 'red', capsize = 10, width = 0.3)
plt.xticks()

plt.ylabel ("Pressure(in)")

plt.subplot(1, 4, 4)
means4 = [np.mean(visibility1), np.mean(visibility2), np.mean(visibility3), np.mean(visibility4)]
stds4 = [np.std(visibility1, ddof = 1), 
        np.std(visibility2, ddof = 1), 
        np.std(visibility3, ddof = 1), 
        np.std(visibility4, ddof = 1)]
plt.bar(attr, means4, yerr = stds4, color = 'purple', capsize = 10, width = 0.3)
plt.xticks()

plt.ylabel ("Visibility(mi)")
#plt.savefig('question1.png')

timezones = [t["Timezone"] for t in data]
tempMean1 = np.mean([d["Temperature(F)"] for d in data if d["Timezone"] == timezones[0]])
tempMean2 = np.mean([d["Temperature(F)"] for d in data if d["Timezone"] == timezones[1]])
tempMean3 = np.mean([d["Temperature(F)"] for d in data if d["Timezone"] == timezones[2]])
tempMean4 = np.mean([d["Temperature(F)"] for d in data if d["Timezone"] == timezones[3]])
tempStd1 = np.std([d["Temperature(F)"] for d in data if d["Timezone"] == timezones[0]])
tempStd2 = np.std([d["Temperature(F)"] for d in data if d["Timezone"] == timezones[1]])
tempStd3 = np.std([d["Temperature(F)"] for d in data if d["Timezone"] == timezones[2]])
tempStd4 = np.std([d["Temperature(F)"] for d in data if d["Timezone"] == timezones[3]])


result = []

for zone in attr:
    r = {
        'TimeZone': zone, 
        'Temperature(F)_mean': np.mean([d["Temperature(F)"] for d in data if d["Timezone"] == zone]),
        'Temperature(F)_std': np.std([d["Temperature(F)"] for d in data if d["Timezone"] == zone]),
        'Humidity(%)_mean': np.mean([d["Humidity(%)"] for d in data if d["Timezone"] == zone]),
        'Humidity(%)_std': np.std([d["Humidity(%)"] for d in data if d["Timezone"] == zone]),
        'Pressure(in)_mean': np.mean([d["Pressure(in)"] for d in data if d["Timezone"] == zone]),
        'Pressure(in)_std': np.std([d["Pressure(in)"] for d in data if d["Timezone"] == zone]),
        'Visibility(mi)_mean': np.mean([d["Visibility(mi)"] for d in data if d["Timezone"] == zone]),
        'Visibility(mi)_std': np.std([d["Visibility(mi)"] for d in data if d["Timezone"] == zone])
    }
    result.append(r)
result

with open("question1.csv", encoding = "utf-8", mode = 'w') as csv_result:
    csv_writer = csv.writer(csv_result, 
                            delimiter = ",", # seperates every value with a comma
                            quotechar = '"', # puts a quote around multiple same ones
                            quoting = csv.QUOTE_MINIMAL, 
                            lineterminator = "\n")
    csv_writer.writerow(result[0].keys())
    for d in result:
        csv_writer.writerow(d.values())
```

```python
# Q2

from collections import Counter
from collections import defaultdict

def q2_visual(ps):
    colors = ["blue", "orange", "green"]
    correlations = []
    plt.figure(figsize = (10, 8))
    plt.subplot(1, 1, 1) # first # - row, second # - column, third # - order in which it is shown
    for i, (state_name, accident_count) in enumerate(ps.items()):
        th_element = [(d["Temperature(F)"], d["Humidity(%)"]) for d in data if d["State"] == state_name]
        corr_mtx = np.corrcoef([d[0] for d in th_element], [d[1] for d in th_element])
        plt.scatter([g[0] for g in th_element], [g[1] for g in th_element], s = 100, c = colors[i], marker = "+", alpha = 0.5, label = state_name)
        correlations.append(corr_mtx)
    return correlations
    plt.legend()

    plt.xlabel ("Temperature(F)")
    plt.ylabel ("Humidity(%)")
    #plt.show()
    plt.savefig("Q2.png")

# Get all the states and put them in a list
states = [d["State"] for d in data]

# Create a (state:count) dictionary
state_dict = Counter(states)

# popular_states will be a dict of 3 top states with their accident count numbers
popular_states = {k: v for k, v in sorted(state_dict.items(), key = lambda item: item[1], reverse = True)[:3]}

print(popular_states)

# calling the function that plots the graph for us, 
#outputs correlation values for the top 3 states as well
correlation = q2_visual(popular_states)
#print(correlation)

# create a list that will hold our dictionary
statesdict_list = []
for i, (k, v) in enumerate(popular_states.items()):
    states_dict = {"State": k, "Accident": v, "Correlation": correlation[i][0][1]}
    statesdict_list.append(states_dict)
    
statesdict_list

with open("question2.csv", encoding = "utf-8", mode = 'w') as csv_result:
    csv_writer = csv.writer(csv_result, 
                            delimiter = ",", # seperates every value with a comma
                            quotechar = '"', # puts a quote around multiple same ones
                            quoting = csv.QUOTE_MINIMAL, 
                            lineterminator = "\n")
    csv_writer.writerow(statesdict_list[0].keys())
    for d in statesdict_list:
        csv_writer.writerow(d.values())
```

```python
#Q3 all your answers to Q3 must be only in this cell.
def q3_visual(al):
    state_values = []
    crossing_values = []
    stop_values = []
    traffic_signals = []
    
    for d in al:
        state_values.append(d["State"]) # x
        crossing_values.append(d["Crossing"]) # y1
        stop_values.append(d["Stop"]) # y2
        traffic_signals.append(d["Traffic_Signal"]) # y3
    
    x_values = range(len(state_values))
    
    w = 0.2
    plt.figure(figsize = (20, 8))
    plt.bar([x - 0.8 * w for x in x_values], crossing_values, color = "blue", width = w, label = "Crossing")
    plt.bar([x + 0.2 * w for x in x_values], stop_values, color = "orange", width = w, label = "Stop")
    plt.bar([x + 1.2 * w for x in x_values], traffic_signals, color = "green", width = w, label = "Traffic_Signal")

    x_ticks = state_values
    plt.xticks(x_values, x_ticks)
    plt.ylim(0, 100)
    plt.legend(loc = "best")
    plt.yticks(np.arange(0, 120, 20))
    plt.xlabel("States")
    plt.ylabel("Percentage")
    #plt.show()

    # Saving the graph
    plt.savefig('question3.png')

from collections import Counter

# Get all the states and put them in a list
states = [d["State"] for d in data]
state_dict = Counter(states)

state_accident = {k: v for k, v in sorted(state_dict.items(), key = lambda item: item[0], reverse = False)}

state_accident

# building state:accident dictionary
accidents_list = []
for state_name, state_count in state_accident.items():
    crossings = len([d["Crossing"] for d in data if d["Crossing"] == "True" and d["State"] == state_name])/state_count 
    stop = len([d["Stop"] for d in data if d["Stop"] == "True" and d["State"] == state_name])/state_count
    traffic_signals = len([d["Traffic_Signal"] for d in data if d["Traffic_Signal"] == "True" and d["State"] == state_name])/state_count
    
    accidents_prob = {"State": state_name, "Crossing": crossings * 100, "Stop": stop * 100, "Traffic_Signal": traffic_signals * 100}
    accidents_list.append(accidents_prob)

accidents_list

# Calling Visualization graph
q3_visual(accidents_list)

with open("question3.csv", encoding = "utf-8", mode = 'w') as csv_result:
    csv_writer = csv.writer(csv_result, 
                            delimiter = ",", # seperates every value with a comma
                            quotechar = '"', # puts a quote around multiple same ones
                            quoting = csv.QUOTE_MINIMAL, 
                            lineterminator = "\n")
    csv_writer.writerow(accidents_list[0].keys())
    for d in accidents_list:
        csv_writer.writerow(d.values())
```

```python
#Q4 all your answers to Q4 must be only in this cell.

def q4_visual():
    
    severity1 = [d for d in data if d["Severity"] == 1]
    severity2 = [d for d in data if d["Severity"] == 2]
    severity3 = [d for d in data if d["Severity"] == 3]
    severity4 = [d for d in data if d["Severity"] == 4]

    plt.figure(figsize = (18, 4))

    temp_dict = [{"Severity": severity1, "Color": 'b-'},
                 {"Severity": severity2, "Color": 'orange'}, 
                 {"Severity": severity3, "Color": 'g-'}, 
                 {"Severity": severity4, "Color": 'r-'}]

    plt.subplot(1, 3, 1)
    xs = np.arange(70, 110, 0.5)
    for td in temp_dict: 
        mean = np.mean([d["Temperature(F)"] for d in td["Severity"]])
        std = np.std([d["Temperature(F)"] for d in td["Severity"]], ddof = 1)
        plt.plot(xs, norm.pdf(xs, mean, std), td["Color"])
        plt.xlabel("States")
    plt.ylabel("Percentage")

    plt.subplot(1, 3, 2)
    xs = np.arange(20, 80, 1)
    for td in temp_dict: 
        mean = np.mean([d["Humidity(%)"] for d in td["Severity"]])
        std = np.std([d["Humidity(%)"] for d in td["Severity"]], ddof = 1)
        plt.plot(xs, norm.pdf(xs, mean, std), td["Color"])

    plt.subplot(1, 3, 3)
    xs = np.arange(0, 10, 0.1)
    for td in temp_dict: 
        mean = np.mean([d["Visibility(mi)"] for d in td["Severity"]])
        std = np.std([d["Visibility(mi)"] for d in td["Severity"]], ddof = 1)
        plt.plot(xs, norm.pdf(xs, mean, std), td["Color"])
    
    #plt.show()
    plt.savefig('question4Test.png')
    
# calling the visualization function
q4_visual()


# with open('question4.csv',encoding="utf-8", mode='w') as csv_result:
    
```
